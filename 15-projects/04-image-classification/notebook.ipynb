## Project Overview
**Goal**: Classify images using a CNN.

**Skills demonstrated**: PyTorch CNN, training loop, evaluation.

```python
import torch
import torch.nn as nn
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np

transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,0.5,0.5),(0.5,0.5,0.5))])

train_set = torchvision.datasets.CIFAR10('./data', train=True, download=True, transform=transform)
test_set = torchvision.datasets.CIFAR10('./data', train=False, download=True, transform=transform)
train_loader = DataLoader(train_set, batch_size=128, shuffle=True)
test_loader = DataLoader(test_set, batch_size=128)

classes = ('plane','car','bird','cat','deer','dog','frog','horse','ship','truck')
print('Classes:', classes)
```

## CNN Architecture

```python
class CNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3,32,3,padding=1), nn.BatchNorm2d(32), nn.ReLU(), nn.MaxPool2d(2),
            nn.Conv2d(32,64,3,padding=1), nn.BatchNorm2d(64), nn.ReLU(), nn.MaxPool2d(2),
        )
        self.classifier = nn.Sequential(
            nn.Flatten(), nn.Linear(64*8*8, 256), nn.ReLU(), nn.Dropout(0.4), nn.Linear(256,10)
        )
    def forward(self, x): return self.classifier(self.features(x))

model = CNN()
print(f'Parameters: {sum(p.numel() for p in model.parameters()):,}')
```

## Training
Note: Run this on GPU for faster training (`device = torch.device('cuda')`).

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.5)

print(f'Training on: {device}')
print('Run training loop here (5-10 epochs recommended)')
print('Expected accuracy: ~65-70% after 10 epochs')
```
