## ML Bias — What Can Go Wrong?
Models can perpetuate and amplify societal biases present in training data.

## Types of Bias
- **Historical bias**: Training data reflects past discrimination
- **Representation bias**: Some groups underrepresented in data
- **Measurement bias**: Features measured differently across groups
- **Aggregation bias**: One model for groups with different patterns

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Synthetic hiring dataset with bias
np.random.seed(42)
n = 1000
df = pd.DataFrame({
    'skill_score': np.random.randn(n) + 2,
    'gender': np.random.choice([0, 1], n),  # 0=male, 1=female
    'hired': np.zeros(n)
})
# Inject historical bias: female candidates need higher scores
df.loc[df['gender']==0, 'hired'] = (df[df['gender']==0]['skill_score'] > 1.5).astype(int)
df.loc[df['gender']==1, 'hired'] = (df[df['gender']==1]['skill_score'] > 2.5).astype(int)

model = LogisticRegression()
model.fit(df[['skill_score','gender']], df['hired'])

for gender, name in [(0,'Male'), (1,'Female')]:
    subset = df[df['gender']==gender]
    acc = accuracy_score(subset['hired'], model.predict(subset[['skill_score','gender']]))
    hire_rate = model.predict(subset[['skill_score','gender']]).mean()
    print(f'{name}: Accuracy={acc:.3f}, Predicted hire rate={hire_rate:.3f}')
```

## Fairness Metrics
- **Demographic parity**: Equal hire rate across groups
- **Equal opportunity**: Equal true positive rate across groups
- **Equalized odds**: Equal TPR and FPR across groups

**No perfect fairness metric exists** — they are mathematically incompatible in most situations.
