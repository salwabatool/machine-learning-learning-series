## Production ML Checklist

Before deploying any ML model, go through this checklist.

## 1. Data Quality
- [ ] Check for missing values and handle appropriately
- [ ] Check for class imbalance — stratify if needed
- [ ] Verify no data leakage between train/val/test
- [ ] Check temporal splits for time-series data
- [ ] Validate feature distributions make business sense

## 2. Model Development
- [ ] Establish a simple baseline first
- [ ] Use cross-validation for robust evaluation
- [ ] Check learning curves for over/underfitting
- [ ] Tune hyperparameters systematically (GridSearch/Optuna)
- [ ] Test multiple model types before choosing

## 3. Evaluation
- [ ] Choose metrics aligned with business goals (not just accuracy)
- [ ] Evaluate on holdout test set only once
- [ ] Check performance across subgroups (fairness)
- [ ] Calculate confidence intervals for metrics

## 4. Deployment
- [ ] Serialize model with versioning
- [ ] Write input validation for the API
- [ ] Add logging for predictions
- [ ] Set up monitoring for data drift
- [ ] Plan for retraining schedule

```python
# Baseline comparison function
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import cross_val_score
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

X, y = load_breast_cancer(return_X_y=True)

models = [
    ('Majority Class Baseline', DummyClassifier(strategy='most_frequent')),
    ('Logistic Regression', LogisticRegression(max_iter=1000)),
    ('Random Forest', RandomForestClassifier(100, random_state=42))
]

for name, model in models:
    score = cross_val_score(model, X, y, cv=5, scoring='f1').mean()
    print(f'{name:30s}: F1 = {score:.4f}')
```
