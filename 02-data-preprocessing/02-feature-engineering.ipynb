## What is Feature Engineering?
Creating new, informative features from existing ones to improve model performance.

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'date': pd.date_range('2023-01-01', periods=5),
    'price': [100, 150, 130, 200, 180],
    'quantity': [10, 5, 8, 3, 6]
})
print(df)
```

```python
# Extract datetime features
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day_of_week'] = df['date'].dt.dayofweek

# Create interaction feature
df['revenue'] = df['price'] * df['quantity']

# Log transform (reduces skewness)
df['log_price'] = np.log1p(df['price'])

print(df.head())
```

## Binning / Discretization

```python
ages = [15, 22, 35, 48, 62, 75]
bins = [0, 18, 35, 55, 100]
labels = ['Youth', 'Young Adult', 'Middle-aged', 'Senior']
result = pd.cut(ages, bins=bins, labels=labels)
print(result)
```
