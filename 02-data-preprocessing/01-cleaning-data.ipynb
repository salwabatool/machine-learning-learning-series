## Why Data Cleaning?
Real-world data is messy — nulls, duplicates, outliers. Garbage in = garbage out.

```python
import pandas as pd
import numpy as np

np.random.seed(42)
df = pd.DataFrame({
    'age': [25, None, 35, 35, -5, 200, 30],
    'income': [50000, 60000, None, 75000, 80000, 90000, 60000],
    'city': ['NY', 'LA', 'NY', 'NY', 'LA', None, 'LA']
})
print(df)
```

```python
# Check missing values
print('\nMissing values:')
print(df.isnull().sum())
```

```python
# Handle missing values
df['age'].fillna(df['age'].median(), inplace=True)
df['income'].fillna(df['income'].mean(), inplace=True)
df['city'].fillna(df['city'].mode()[0], inplace=True)
print('\nAfter filling missing values:')
print(df.isnull().sum())
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)
print('\nAfter removing duplicates:', df.shape)
```

```python
# Remove outliers using IQR
Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['age'] >= Q1 - 1.5*IQR) & (df['age'] <= Q3 + 1.5*IQR)]
print('\nAfter outlier removal:', df.shape)
```
