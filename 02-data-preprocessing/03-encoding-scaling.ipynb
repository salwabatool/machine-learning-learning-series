# Encoding & Scaling

> Part of the **Machine Learning Learning Series**

## Why Encoding & Scaling?
- ML models need numbers — categorical features must be encoded
- Many algorithms are sensitive to feature scale

## Categorical Encoding

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

df = pd.DataFrame({'color': ['red','blue','green','red','blue']})

# Label Encoding (for ordinal categories)
le = LabelEncoder()
df['color_label'] = le.fit_transform(df['color'])

# One-Hot Encoding (for nominal categories)
df_encoded = pd.get_dummies(df['color'], prefix='color')
print(df.head())
print(df_encoded.head())
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import numpy as np

data = np.array([[1000, 2], [2000, 5], [1500, 3], [3000, 8]])

# StandardScaler: mean=0, std=1
ss = StandardScaler()
print('Standard Scaled:')
print(ss.fit_transform(data))

# MinMaxScaler: range [0,1]
mm = MinMaxScaler()
print('\nMinMax Scaled:')
print(mm.fit_transform(data))
```
