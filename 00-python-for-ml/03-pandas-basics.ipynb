## What is Pandas?
Pandas provides DataFrame — a 2D table structure ideal for loading, exploring, and manipulating datasets.

```python
import pandas as pd
import numpy as np

# Create a DataFrame
df = pd.DataFrame({
    'age': [25, 30, 35, 40],
    'salary': [50000, 60000, 75000, 90000],
    'department': ['HR', 'Eng', 'Eng', 'HR']
})
print(df)
```

## Exploring Data

```python
print(df.shape)
print(df.dtypes)
print(df.describe())
print(df.head(2))
```

## Selection & Filtering

```python
# Select columns
print(df['age'])
# Filter rows
print(df[df['salary'] > 60000])
```

## Handling Missing Values

```python
df2 = df.copy()
df2.loc[1, 'salary'] = None
print('Missing:', df2.isnull().sum())
df2['salary'].fillna(df2['salary'].mean(), inplace=True)
print('After fill:', df2.isnull().sum())
```

## GroupBy

```python
print(df.groupby('department')['salary'].mean())
```
