# Data Visualization for ML

> Part of the **Machine Learning Learning Series**

## Why Visualize?
Visualization helps understand data distributions, relationships, and anomalies before modeling.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
df = pd.DataFrame({
    'feature1': np.random.randn(200),
    'feature2': np.random.randn(200) * 2 + 1,
    'label': np.random.choice(['A','B'], 200)
})
```

## Histogram — Distribution

```python
plt.figure(figsize=(8,4))
df['feature1'].hist(bins=30, color='steelblue', edgecolor='white')
plt.title('Feature 1 Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

## Scatter Plot — Relationships

```python
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x='feature1', y='feature2', hue='label')
plt.title('Feature1 vs Feature2')
plt.show()
```

## Correlation Heatmap

```python
plt.figure(figsize=(5,4))
sns.heatmap(df[['feature1','feature2']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

## Boxplot — Outliers

```python
plt.figure(figsize=(8,4))
sns.boxplot(data=df, x='label', y='feature1')
plt.title('Feature1 by Label')
plt.show()
```

## Summary
- Always visualize before modeling
- Histograms, scatter plots, heatmaps, and boxplots cover 90% of EDA needs
