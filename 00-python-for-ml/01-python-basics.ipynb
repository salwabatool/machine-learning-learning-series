## Why Python for ML?
Python is the most popular language for ML due to its simplicity, rich ecosystem, and community support.

```python
# Variables and data types
x = 10
name = 'ML'
pi = 3.14
is_ml = True
print(x, name, pi, is_ml)
```

## Control Flow

```python
# Loops and conditionals
for i in range(5):
    if i % 2 == 0:
        print(f'{i} is even')
    else:
        print(f'{i} is odd')
```

## Functions

```python
def greet(name='World'):
    return f'Hello, {name}!'

print(greet('ML Learner'))
```

## List Comprehensions (very useful in ML)

```python
squares = [x**2 for x in range(10)]
print(squares)
```
