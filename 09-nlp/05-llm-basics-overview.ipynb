## Large Language Models (LLMs)
Massively scaled transformer models trained on internet-scale text data.

## Evolution of NLP
```
Bag of Words → Word2Vec → LSTM → Transformer → BERT → GPT → LLMs (GPT-4, Claude, Llama)
```

## Key Concepts
### Pre-training
Train on massive text corpus to learn language understanding.
- **BERT**: Masked language modeling (bidirectional)
- **GPT**: Next token prediction (autoregressive)

### Fine-tuning
Adapt pre-trained model to specific tasks with smaller labeled datasets.

````python
# Using HuggingFace transformers
try:
    from transformers import pipeline
    
    # Sentiment analysis with pre-trained model
    classifier = pipeline('sentiment-analysis', model='distilbert-base-uncased-finetuned-sst-2-english')
    results = classifier(['Machine learning is fantastic!', 'This model is terrible.'])
    for r in results:
        print(f"Label: {r['label']}, Score: {r['score']:.4f}")
except ImportError:
    print('Install: pip install transformers')
    print('Example output:')
    print("POSITIVE, Score: 0.9998")
    print("NEGATIVE, Score: 0.9995")
````

## Prompting LLMs
- **Zero-shot**: Direct question, no examples
- **Few-shot**: Provide 2-3 examples in the prompt
- **Chain-of-thought**: Ask model to reason step by step
- **RAG**: Retrieval Augmented Generation — give model relevant docs
