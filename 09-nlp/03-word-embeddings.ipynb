## Word Embeddings
Dense vector representations where similar words have similar vectors.

'king' - 'man' + 'woman' ≈ 'queen'

```python
# Using pre-trained embeddings via gensim
try:
    from gensim.models import Word2Vec
    
    sentences = [
        ['machine', 'learning', 'is', 'great'],
        ['deep', 'learning', 'uses', 'neural', 'networks'],
        ['neural', 'networks', 'are', 'powerful'],
        ['machine', 'learning', 'and', 'deep', 'learning'],
        ['data', 'science', 'involves', 'machine', 'learning']
    ]
    
    model = Word2Vec(sentences, vector_size=50, window=3, min_count=1, epochs=100)
    print('Vocabulary:', list(model.wv.key_to_index.keys()))
    print('\nSimilar to learning:')
    print(model.wv.most_similar('learning', topn=3))
except ImportError:
    print('Install gensim: pip install gensim')
```

## Types of Embeddings
| Method | Description |
|--------|-------------|
| Word2Vec | Predict word from context (CBOW) or context from word (Skip-gram) |
| GloVe | Global co-occurrence statistics |
| FastText | Subword embeddings — handles OOV words |
| BERT | Contextual embeddings — same word, different meanings |
