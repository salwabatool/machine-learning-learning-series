## Text Preprocessing Pipeline
Raw text → Clean tokens ready for ML models.

```python
import re
import nltk
nltk.download('punkt', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download('wordnet', quiet=True)
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

text = 'The quick brown foxes are jumping over the lazy dogs! NLP is AMAZING :)'
print('Original:', text)
```

```python
# Lowercase
text_lower = text.lower()
print('Lowercase:', text_lower)

# Remove special characters
text_clean = re.sub(r'[^a-z\s]', '', text_lower)
print('Clean:', text_clean)

# Tokenize
tokens = word_tokenize(text_clean)
print('Tokens:', tokens)

# Remove stopwords
stop_words = set(stopwords.words('english'))
tokens_filtered = [t for t in tokens if t not in stop_words]
print('No stopwords:', tokens_filtered)

# Lemmatize
lemmatizer = WordNetLemmatizer()
tokens_lem = [lemmatizer.lemmatize(t) for t in tokens_filtered]
print('Lemmatized:', tokens_lem)
```
