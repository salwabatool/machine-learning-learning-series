## Text Classification Pipeline
TF-IDF + ML classifier = fast, effective text classifier.

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, accuracy_score

categories = ['sci.space', 'rec.sport.hockey', 'talk.politics.misc', 'comp.graphics']
train = fetch_20newsgroups(subset='train', categories=categories, remove=('headers','footers','quotes'))
test = fetch_20newsgroups(subset='test', categories=categories, remove=('headers','footers','quotes'))
```

```python
# Pipeline 1: TF-IDF + Naive Bayes
nb_pipeline = Pipeline([('tfidf', TfidfVectorizer(max_features=5000, ngram_range=(1,2))),
                        ('clf', MultinomialNB())])
nb_pipeline.fit(train.data, train.target)
print(f'Naive Bayes: {accuracy_score(test.target, nb_pipeline.predict(test.data)):.4f}')
```

```python
# Pipeline 2: TF-IDF + Logistic Regression
lr_pipeline = Pipeline([('tfidf', TfidfVectorizer(max_features=5000, ngram_range=(1,2))),
                        ('clf', LogisticRegression(max_iter=1000))])
lr_pipeline.fit(train.data, train.target)
y_pred = lr_pipeline.predict(test.data)
print(f'Logistic Regression: {accuracy_score(test.target, y_pred):.4f}')
print(classification_report(test.target, y_pred, target_names=categories))
```
