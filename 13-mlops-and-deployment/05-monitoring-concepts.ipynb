## Why Monitor ML Models?
Models degrade over time as the real world changes (data drift, concept drift).

## Types of Drift

| Type | Description |
|------|-------------|
| **Data drift** | Input distribution changes |
| **Label drift** | Output distribution changes |
| **Concept drift** | Relationship between X and y changes |
| **Feature drift** | Individual feature statistics shift |

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ks_2samp

# Simulate data drift
np.random.seed(42)
train_dist = np.random.normal(50, 10, 1000)  # Training distribution
prod_dist = np.random.normal(58, 12, 1000)   # Production (drifted)

stat, p_value = ks_2samp(train_dist, prod_dist)
print(f'KS Statistic: {stat:.4f}')
print(f'P-value: {p_value:.6f}')
print('Drift detected!' if p_value < 0.05 else 'No significant drift')
```

```python
plt.figure(figsize=(8,4))
plt.hist(train_dist, bins=40, alpha=0.5, label='Training', color='blue')
plt.hist(prod_dist, bins=40, alpha=0.5, label='Production', color='red')
plt.title('Data Drift: Training vs Production')
plt.legend()
plt.show()
```
