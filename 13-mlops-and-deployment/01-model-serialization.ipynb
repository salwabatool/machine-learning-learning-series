## Saving & Loading Models
After training, persist models to disk for later use or deployment.

```python
import joblib
import pickle
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

model = RandomForestClassifier(n_estimators=50, random_state=42)
model.fit(X_train, y_train)
print(f'Model accuracy: {model.score(X_test, y_test):.4f}')
```

```python
# Save with joblib (recommended for sklearn)
joblib.dump(model, '/tmp/rf_model.joblib')
loaded_model = joblib.load('/tmp/rf_model.joblib')
print(f'Loaded model accuracy: {loaded_model.score(X_test, y_test):.4f}')
```

```python
# Save with pickle
with open('/tmp/rf_model.pkl', 'wb') as f:
    pickle.dump(model, f)

with open('/tmp/rf_model.pkl', 'rb') as f:
    pkl_model = pickle.load(f)
print(f'Pickle model accuracy: {pkl_model.score(X_test, y_test):.4f}')
```

## PyTorch Model Saving

```python
import torch
import torch.nn as nn

model_torch = nn.Linear(4, 3)
torch.save(model_torch.state_dict(), '/tmp/model.pt')

# Load
loaded = nn.Linear(4, 3)
loaded.load_state_dict(torch.load('/tmp/model.pt'))
print('PyTorch model saved and loaded successfully')
```
