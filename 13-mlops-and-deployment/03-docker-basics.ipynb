## Why Docker for ML?
Docker packages your model + dependencies + code into a portable container that runs anywhere.

## Dockerfile Example
```dockerfile
FROM python:3.10-slim

WORKDIR /app

COPY requirements.txt .
RUN pip install --no-cache-dir -r requirements.txt

COPY . .

EXPOSE 8000
CMD ["uvicorn", "app:app", "--host", "0.0.0.0", "--port", "8000"]
```

## Docker Commands
```bash
# Build image
docker build -t ml-api .

# Run container
docker run -p 8000:8000 ml-api

# Run in background
docker run -d -p 8000:8000 --name ml-api ml-api

# Check logs
docker logs ml-api

# Stop container
docker stop ml-api
```

````python
# docker-compose.yml example
docker_compose = '''
version: '3.8'
services:
  ml-api:
    build: .
    ports:
      - "8000:8000"
    volumes:
      - ./models:/app/models
    environment:
      - MODEL_PATH=/app/models/rf_model.joblib
'''
print(docker_compose)
````
