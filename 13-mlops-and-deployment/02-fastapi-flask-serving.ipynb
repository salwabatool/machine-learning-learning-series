## Serving ML Models via API
Wrap your model in a REST API so other services can make predictions.

## FastAPI Example
FastAPI is the recommended modern framework — fast, type-safe, auto-docs.

````python
# app.py — run with: uvicorn app:app --reload
fastapi_code = '''
from fastapi import FastAPI
from pydantic import BaseModel
import joblib
import numpy as np

app = FastAPI(title="ML Model API")
model = joblib.load("rf_model.joblib")

class PredictionRequest(BaseModel):
    features: list[float]

class PredictionResponse(BaseModel):
    prediction: int
    probability: float

@app.get("/")
def root():
    return {"message": "ML API is running"}

@app.post("/predict", response_model=PredictionResponse)
def predict(request: PredictionRequest):
    X = np.array(request.features).reshape(1, -1)
    pred = model.predict(X)[0]
    prob = model.predict_proba(X).max()
    return PredictionResponse(prediction=int(pred), probability=float(prob))
'''

print('FastAPI app code:')
print(fastapi_code)
````

## Testing the API
```bash

uvicorn app:app --reload
```
