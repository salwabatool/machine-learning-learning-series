## Collaborative Filtering
'Users who liked what you liked also liked...' — find similar users or items.

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

np.random.seed(42)
n_users, n_items = 20, 10
ratings_matrix = np.random.randint(0, 6, (n_users, n_items)).astype(float)
ratings_matrix[ratings_matrix == 0] = np.nan
ratings_df = pd.DataFrame(ratings_matrix, columns=[f'item_{i}' for i in range(n_items)])
print(ratings_df.head())
```

```python
# User-based CF
ratings_filled = ratings_df.fillna(ratings_df.mean())
user_similarity = cosine_similarity(ratings_filled)
user_sim_df = pd.DataFrame(user_similarity)

# Recommend for user 0
user_id = 0
similarities = user_sim_df[user_id].drop(user_id).sort_values(ascending=False)
top_similar_users = similarities.head(5).index

# Items user 0 hasn't rated
unrated = ratings_df.iloc[user_id][ratings_df.iloc[user_id].isna()].index

# Average rating from similar users
recommendations = {}
for item in unrated:
    ratings_from_similar = ratings_df.iloc[top_similar_users][item].dropna()
    if len(ratings_from_similar) > 0:
        recommendations[item] = ratings_from_similar.mean()

recs = pd.Series(recommendations).sort_values(ascending=False)
print('\nRecommendations for User 0:')
print(recs)
```
