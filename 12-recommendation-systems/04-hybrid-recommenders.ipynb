## Hybrid Recommender Systems
Combine collaborative and content-based filtering to overcome limitations of each.

## Common Hybrid Strategies

1. **Weighted**: Average scores from multiple recommenders
2. **Switching**: Use CF when data available, else content-based
3. **Feature combination**: Feed content features into CF model
4. **Cascade**: Use one filter to produce candidates, another to rank

```python
import numpy as np
import pandas as pd

# Example: weighted hybrid
def hybrid_recommend(user_id, cf_scores, cb_scores, alpha=0.6):
    """
    alpha: weight for collaborative filtering (1-alpha for content-based)
    """
    hybrid = alpha * cf_scores + (1 - alpha) * cb_scores
    return np.argsort(hybrid)[::-1]

# Simulate scores for 10 items
np.random.seed(42)
cf_scores = np.random.rand(10)   # from CF model
cb_scores = np.random.rand(10)   # from content-based model

ranking = hybrid_recommend(user_id=0, cf_scores=cf_scores, cb_scores=cb_scores, alpha=0.6)
print('Hybrid recommendation ranking (item indices):')
print(ranking[:5])
```
