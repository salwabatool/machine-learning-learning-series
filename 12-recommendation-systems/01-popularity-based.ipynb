## Simplest Recommender: Popularity
Recommend the most popular items — no personalization needed.

```python
import pandas as pd
import numpy as np

np.random.seed(42)
n_users, n_items = 100, 50
ratings = pd.DataFrame({
    'user_id': np.random.randint(0, n_users, 500),
    'item_id': np.random.randint(0, n_items, 500),
    'rating': np.random.choice([1,2,3,4,5], 500, p=[0.05,0.1,0.2,0.35,0.3])
}).drop_duplicates(['user_id','item_id'])
print(f'Ratings shape: {ratings.shape}')
```

```python
# Popularity score: average rating + count weighting
stats = ratings.groupby('item_id')['rating'].agg(['mean','count']).reset_index()
stats.columns = ['item_id','avg_rating','num_ratings']
stats['popularity_score'] = stats['avg_rating'] * np.log1p(stats['num_ratings'])
top10 = stats.nlargest(10, 'popularity_score')
print('Top 10 most popular items:')
print(top10)
```

## Limitations
- No personalization — same recommendation for everyone
- Cold start: struggles with new items
- Use as a fallback when user history is unavailable
