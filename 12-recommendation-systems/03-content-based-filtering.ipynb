## Content-Based Filtering
Recommend items similar to what the user has liked based on item features.

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Movie dataset example
movies = pd.DataFrame({
    'title': ['The Matrix', 'Inception', 'Interstellar', 'The Dark Knight', 'Avengers',
              'Iron Man', 'Gravity', 'Blade Runner', 'Ex Machina', 'Her'],
    'genres': ['sci-fi action thriller', 'sci-fi thriller mystery',
               'sci-fi drama space', 'action crime thriller',
               'action superhero sci-fi', 'action superhero comedy',
               'sci-fi thriller space', 'sci-fi noir dystopia',
               'sci-fi thriller ai', 'sci-fi romance drama']
})
```

```python
# TF-IDF on genres
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(movies['genres'])
sim_matrix = cosine_similarity(tfidf_matrix)

# Get recommendations for a movie
def recommend(title, n=5):
    idx = movies[movies['title'] == title].index[0]
    scores = list(enumerate(sim_matrix[idx]))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)[1:n+1]
    return [movies['title'].iloc[i] for i, _ in scores]

print('Recommendations for "The Matrix":')
for r in recommend('The Matrix'):
    print(f'  - {r}')
```
