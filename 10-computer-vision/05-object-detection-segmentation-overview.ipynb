## Beyond Classification

### Object Detection
Locate AND classify multiple objects in an image (bounding boxes).

**Key Models**: YOLO, SSD, Faster R-CNN, DETR

### Image Segmentation
Assign a class label to every pixel in the image.

- **Semantic segmentation**: Label each pixel (no instance distinction)
- **Instance segmentation**: Separate each object instance
- **Panoptic segmentation**: Combines both

## Architecture Summary
| Task | Input | Output | Example Model |
|------|-------|--------|---------------|
| Classification | Image | 1 label | ResNet, VGG |
| Detection | Image | Boxes + labels | YOLO, DETR |
| Segmentation | Image | Pixel masks | U-Net, Mask R-CNN |
