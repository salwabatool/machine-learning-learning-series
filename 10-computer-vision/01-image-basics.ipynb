## Digital Images
Images are 3D arrays: Height × Width × Channels (RGB = 3 channels).

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a simple synthetic image
img = np.zeros((100, 100, 3), dtype=np.uint8)
img[20:80, 20:80] = [255, 0, 0]   # Red square
img[40:60, 40:60] = [0, 0, 255]   # Blue square

plt.figure(figsize=(8,3))
plt.subplot(131); plt.imshow(img); plt.title('RGB'); plt.axis('off')
plt.subplot(132); plt.imshow(img[:,:,0], cmap='Reds'); plt.title('R channel'); plt.axis('off')
plt.subplot(133); plt.imshow(img[:,:,2], cmap='Blues'); plt.title('B channel'); plt.axis('off')
plt.show()
print(f'Image shape: {img.shape}, dtype: {img.dtype}')
```

## Grayscale Conversion

```python
gray = np.mean(img, axis=2).astype(np.uint8)
plt.figure(figsize=(5,4))
plt.imshow(gray, cmap='gray')
plt.title('Grayscale')
plt.colorbar()
plt.show()
```
