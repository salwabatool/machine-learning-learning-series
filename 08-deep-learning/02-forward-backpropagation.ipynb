# Forward & Backpropagation

> Part of the **Machine Learning Learning Series**

## Forward Pass
Compute predictions layer by layer: Input → Hidden → Output

```python
import numpy as np

# Simple 2-layer network from scratch
np.random.seed(42)

# Network architecture: 2 inputs → 3 hidden → 1 output
W1 = np.random.randn(2, 3) * 0.1
b1 = np.zeros((1, 3))
W2 = np.random.randn(3, 1) * 0.1
b2 = np.zeros((1, 1))

def sigmoid(x): return 1 / (1 + np.exp(-x))
def sigmoid_derivative(x): return sigmoid(x) * (1 - sigmoid(x))

# XOR dataset
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])
```

```python
# Training loop
for epoch in range(5000):
    # Forward pass
    z1 = X @ W1 + b1
    a1 = sigmoid(z1)
    z2 = a1 @ W2 + b2
    a2 = sigmoid(z2)
    
    # Loss
    loss = np.mean((y - a2)**2)
    
    # Backward pass
    dL_da2 = -2 * (y - a2)
    da2_dz2 = sigmoid_derivative(z2)
    dL_dz2 = dL_da2 * da2_dz2
    
    W2 -= 0.5 * a1.T @ dL_dz2
    b2 -= 0.5 * np.sum(dL_dz2, axis=0, keepdims=True)
    
    dL_da1 = dL_dz2 @ W2.T
    dL_dz1 = dL_da1 * sigmoid_derivative(z1)
    
    W1 -= 0.5 * X.T @ dL_dz1
    b1 -= 0.5 * np.sum(dL_dz1, axis=0, keepdims=True)

print('Predictions after training:')
print(np.round(a2, 3))
print('Targets:', y.ravel())
```
