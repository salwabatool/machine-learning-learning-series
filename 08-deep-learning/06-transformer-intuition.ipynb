## What is the Transformer?
Architecture based on self-attention — processes entire sequences in parallel rather than step-by-step.

## Key Concepts
### Self-Attention
Each token attends to all other tokens, learning which are most relevant.

Attention(Q,K,V) = softmax(QK^T / √d_k) * V

````python
import torch
import torch.nn.functional as F
import numpy as np

# Simple self-attention example
def self_attention(Q, K, V):
    d_k = Q.shape[-1]
    scores = Q @ K.transpose(-2, -1) / np.sqrt(d_k)
    weights = F.softmax(scores, dim=-1)
    return weights @ V, weights

# Example: 4 tokens, 8-dim embeddings
torch.manual_seed(42)
Q = torch.randn(1, 4, 8)  # queries
K = torch.randn(1, 4, 8)  # keys
V = torch.randn(1, 4, 8)  # values

output, attn_weights = self_attention(Q, K, V)
print('Attention weights shape:', attn_weights.shape)
print('Output shape:', output.shape)
````

## Transformer Architecture
```
Input → Embedding + Positional Encoding
↓
Multi-Head Attention → Add & Norm
↓
Feed Forward → Add & Norm
↓
Repeat N times
↓
Output
```

**Famous Transformers**: BERT, GPT, T5, Vision Transformer (ViT)
