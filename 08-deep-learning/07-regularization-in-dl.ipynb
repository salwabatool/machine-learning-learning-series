## Overfitting in Deep Learning
Deep networks have many parameters and can easily overfit. Regularization techniques prevent this.

```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt

# Compare models with/without regularization
class Model(nn.Module):
    def __init__(self, use_dropout=False, use_batchnorm=False):
        super().__init__()
        layers = [nn.Linear(10, 64), nn.ReLU()]
        if use_dropout: layers.append(nn.Dropout(0.5))
        if use_batchnorm: layers.append(nn.BatchNorm1d(64))
        layers += [nn.Linear(64, 64), nn.ReLU(), nn.Linear(64, 1)]
        self.net = nn.Sequential(*layers)
    def forward(self, x): return self.net(x)

print('Models created — no_reg, with_dropout, with_batchnorm')
for name, m in [('No regularization', Model()), ('With Dropout', Model(use_dropout=True)),
                ('With BatchNorm', Model(use_batchnorm=True))]:
    params = sum(p.numel() for p in m.parameters())
    print(f'{name}: {params} parameters')
```

## Regularization Techniques
| Technique | What it does |
|-----------|--------------|
| **Dropout** | Randomly zeros out neurons during training |
| **Batch Normalization** | Normalizes layer outputs — faster training |
| **L2 Weight Decay** | Penalizes large weights (via optimizer) |
| **Early Stopping** | Stop when validation loss stops improving |
| **Data Augmentation** | Create more varied training samples |
