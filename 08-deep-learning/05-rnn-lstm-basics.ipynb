## Recurrent Neural Networks
Designed for sequential data — hidden state carries information across time steps.

```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt

# Generate sine wave
t = np.linspace(0, 4*np.pi, 200)
y = np.sin(t)

# Create sequences
seq_len = 20
X_seq, y_seq = [], []
for i in range(len(y) - seq_len):
    X_seq.append(y[i:i+seq_len])
    y_seq.append(y[i+seq_len])

X_seq = torch.FloatTensor(X_seq).unsqueeze(-1)
y_seq = torch.FloatTensor(y_seq).unsqueeze(-1)
print(f'X shape: {X_seq.shape}, y shape: {y_seq.shape}')
```

```python
# LSTM model
class LSTMPredictor(nn.Module):
    def __init__(self, input_size=1, hidden_size=32, num_layers=1):
        super().__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, 1)
    def forward(self, x):
        out, _ = self.lstm(x)
        return self.fc(out[:, -1, :])

model = LSTMPredictor()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
criterion = nn.MSELoss()

for epoch in range(200):
    pred = model(X_seq)
    loss = criterion(pred, y_seq)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
print(f'Final loss: {loss.item():.4f}')
```
