## Filter Methods
Rank features independently of any ML model using statistical measures.

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.feature_selection import SelectKBest, f_classif, chi2
from scipy.stats import pearsonr
import matplotlib.pyplot as plt

X, y = load_breast_cancer(return_X_y=True)
feature_names = load_breast_cancer().feature_names
```

```python
# F-test
selector = SelectKBest(f_classif, k=10)
X_selected = selector.fit_transform(X, y)
scores = pd.Series(selector.scores_, index=feature_names).nlargest(10)
scores.plot(kind='barh', figsize=(8,5))
plt.title('Top 10 Features by F-Score')
plt.show()
```

## Correlation Filter

```python
df = pd.DataFrame(X, columns=feature_names)
corrs = df.corrwith(pd.Series(y)).abs().sort_values(ascending=False)
print(corrs.head(10))
```
