# Embedded Methods for Feature Selection

> Part of the **Machine Learning Learning Series**

## Embedded Methods
Feature selection is built into the model training process.

```python
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import Lasso, LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
import pandas as pd
import matplotlib.pyplot as plt

X, y = load_breast_cancer(return_X_y=True)
feature_names = load_breast_cancer().feature_names
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## Lasso (L1 Regularization) — drives weak features to zero

```python
from sklearn.linear_model import LogisticRegression
lasso_lr = LogisticRegression(penalty='l1', C=0.1, solver='liblinear', random_state=42)
lasso_lr.fit(X_scaled, y)
coef = pd.Series(abs(lasso_lr.coef_[0]), index=feature_names)
selected = coef[coef > 0].sort_values(ascending=False)
print(f'Features selected by Lasso: {len(selected)}')
selected.plot(kind='bar', figsize=(10,4))
plt.title('Lasso Feature Importance')
plt.tight_layout()
plt.show()
```

## Tree-based Feature Importance

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_scaled, y)
coef = pd.Series(rf.feature_importances_, index=feature_names).sort_values(ascending=False)[:10]
coef.plot(kind='barh', figsize=(8,5))
plt.title('Random Forest Feature Importance')
plt.show()
```
