## Practical Feature Engineering Techniques

```python
import pandas as pd
import numpy as np

np.random.seed(42)
df = pd.DataFrame({
    'age': np.random.randint(18, 65, 100),
    'income': np.random.exponential(50000, 100),
    'years_employed': np.random.randint(0, 40, 100),
    'purchases': np.random.poisson(5, 100)
})
```

## Polynomial Features

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(df[['age','income']].values)
print(f'Original: {df[['age','income']].shape} → Polynomial: {X_poly.shape}')
```

## Ratio Features & Domain Knowledge

```python
df['income_per_year'] = df['income'] / (df['years_employed'] + 1)
df['log_income'] = np.log1p(df['income'])
df['age_group'] = pd.cut(df['age'], bins=[18,30,45,65], labels=['young','mid','senior'])
print(df.head())
```
