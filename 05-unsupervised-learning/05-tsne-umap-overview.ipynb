## t-SNE
Non-linear dimensionality reduction optimized for visualization. Preserves local structure.

```python
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

X, y = load_digits(return_X_y=True)

# PCA first to speed up t-SNE
X_pca = PCA(n_components=30).fit_transform(X)
X_tsne = TSNE(n_components=2, random_state=42, perplexity=30).fit_transform(X_pca)

plt.figure(figsize=(8,6))
scatter = plt.scatter(X_tsne[:,0], X_tsne[:,1], c=y, cmap='tab10', alpha=0.7, s=15)
plt.colorbar(scatter)
plt.title('t-SNE: Digits Dataset')
plt.show()
```

## t-SNE vs UMAP
| | t-SNE | UMAP |
