## What is DBSCAN?
Density-based clustering — finds arbitrarily shaped clusters and marks outliers as noise.

```python
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN, KMeans
import matplotlib.pyplot as plt

X, _ = make_moons(n_samples=200, noise=0.1, random_state=42)

fig, axes = plt.subplots(1, 2, figsize=(12,4))
for ax, model, title in zip(axes,
    [KMeans(n_clusters=2, random_state=42), DBSCAN(eps=0.3, min_samples=5)],
    ['K-Means (fails on non-convex)', 'DBSCAN (handles non-convex)']):
    labels = model.fit_predict(X)
    ax.scatter(X[:,0], X[:,1], c=labels, cmap='tab10', alpha=0.8)
    ax.set_title(title)
plt.tight_layout()
plt.show()
```

## Key Parameters
- `eps`: Maximum distance between neighbors
- `min_samples`: Minimum points to form a core point
- Points not in any cluster are labeled -1 (noise/outliers)
