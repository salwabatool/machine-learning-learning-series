# Principal Component Analysis (PCA)

> Part of the **Machine Learning Learning Series**

## What is PCA?
Dimensionality reduction by projecting data onto directions of maximum variance (principal components).

```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

X, y = load_digits(return_X_y=True)
print(f'Original shape: {X.shape}')  # 64 features

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
print(f'Reduced shape: {X_pca.shape}')
print(f'Variance explained: {pca.explained_variance_ratio_.sum():.3f}')
```

```python
plt.figure(figsize=(8,6))
scatter = plt.scatter(X_pca[:,0], X_pca[:,1], c=y, cmap='tab10', alpha=0.7, s=15)
plt.colorbar(scatter)
plt.title('PCA: Digits Dataset (2D)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

## Explained Variance

```python
pca_full = PCA().fit(X)
cumvar = np.cumsum(pca_full.explained_variance_ratio_)
plt.plot(cumvar)
plt.axhline(0.95, color='red', linestyle='--', label='95% variance')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('PCA: Explained Variance')
plt.legend()
plt.show()
```
