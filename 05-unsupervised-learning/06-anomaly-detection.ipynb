## Anomaly Detection
Identify rare items or observations that differ significantly from the majority.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler

np.random.seed(42)
X_normal = np.random.randn(200, 2)
X_outliers = np.random.uniform(-4, 4, (20, 2))
X = np.vstack([X_normal, X_outliers])
y_true = np.hstack([np.ones(200), -np.ones(20)])
```

```python
iso = IsolationForest(contamination=0.1, random_state=42)
y_pred = iso.fit_predict(X)

plt.figure(figsize=(8,6))
plt.scatter(X[y_pred==1, 0], X[y_pred==1, 1], c='blue', label='Normal', alpha=0.5)
plt.scatter(X[y_pred==-1, 0], X[y_pred==-1, 1], c='red', label='Anomaly', s=100, marker='x')
plt.title('Isolation Forest Anomaly Detection')
plt.legend()
plt.show()
```
