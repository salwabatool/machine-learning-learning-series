# Bias-Variance Tradeoff

> Part of the **Machine Learning Learning Series**

## The Bias-Variance Tradeoff
- **High Bias (Underfitting)**: Model too simple, misses patterns
- **High Variance (Overfitting)**: Model too complex, memorizes noise
- **Goal**: Find the sweet spot

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

np.random.seed(42)
X = np.linspace(0, 1, 50).reshape(-1,1)
y = np.sin(2*np.pi*X.ravel()) + np.random.normal(0, 0.1, 50)
X_test = np.linspace(0, 1, 200).reshape(-1,1)
y_test_true = np.sin(2*np.pi*X_test.ravel())

fig, axes = plt.subplots(1, 3, figsize=(15,4))
for ax, degree, title in zip(axes, [1, 4, 15], ['Underfit (d=1)','Good fit (d=4)','Overfit (d=15)']):
    poly = Pipeline([('poly', PolynomialFeatures(degree)), ('lr', LinearRegression())])
    poly.fit(X, y)
    ax.scatter(X, y, alpha=0.6, label='Data')
    ax.plot(X_test, poly.predict(X_test), 'r-', label='Fit')
    ax.set_title(title)
    ax.legend()
plt.tight_layout()
plt.show()
```
