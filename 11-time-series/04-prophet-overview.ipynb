## Facebook Prophet
Additive model for forecasting with trend, seasonality, and holidays.

```python
try:
    from prophet import Prophet
    import pandas as pd
    import numpy as np
    import matplotlib.pyplot as plt
    
    np.random.seed(42)
    dates = pd.date_range('2020-01-01', periods=500, freq='D')
    values = 100 + np.cumsum(np.random.randn(500)*0.3) + 15*np.sin(np.arange(500)*2*np.pi/365)
    df = pd.DataFrame({'ds': dates, 'y': values})
    
    train = df.iloc[:400]
    model = Prophet(seasonality_mode='additive')
    model.fit(train)
    
    future = model.make_future_dataframe(periods=100)
    forecast = model.predict(future)
    model.plot(forecast)
    plt.title('Prophet Forecast')
    plt.show()
except ImportError:
    print('Install: pip install prophet')
    print('Prophet is great for:')
    print('- Business time series with holidays')
    print('- Multiple seasonalities')
    print('- Robust to missing data')
```
