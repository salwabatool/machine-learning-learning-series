## ARIMA Model
AutoRegressive Integrated Moving Average — classical statistical forecasting.

- **AR (p)**: Autoregressive — past values
- **I (d)**: Integrated — differencing for stationarity
- **MA (q)**: Moving average — past forecast errors

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error

np.random.seed(42)
n = 200
ts = np.cumsum(np.random.randn(n)) + 50

train, test = ts[:160], ts[160:]

model = ARIMA(train, order=(2, 1, 2))
result = model.fit()
forecast = result.forecast(steps=40)

plt.figure(figsize=(12,5))
plt.plot(range(160), train, label='Train')
plt.plot(range(160,200), test, label='Actual')
plt.plot(range(160,200), forecast, label='Forecast', linestyle='--')
plt.legend()
plt.title('ARIMA Forecast')
plt.show()
print(f'MAE: {mean_absolute_error(test, forecast):.4f}')
```
