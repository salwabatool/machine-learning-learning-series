## Forecasting as Regression
Create lag features and time-based features, then use any ML model.

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt

np.random.seed(42)
dates = pd.date_range('2020-01-01', periods=365, freq='D')
values = 50 + np.cumsum(np.random.randn(365)*0.5) + 10*np.sin(np.arange(365)*2*np.pi/365)
df = pd.DataFrame({'date': dates, 'value': values})
df.set_index('date', inplace=True)
```

```python
# Create features
for lag in [1, 7, 14, 30]:
    df[f'lag_{lag}'] = df['value'].shift(lag)
df['rolling_7'] = df['value'].rolling(7).mean()
df['rolling_30'] = df['value'].rolling(30).mean()
df['day_of_week'] = df.index.dayofweek
df['month'] = df.index.month
df.dropna(inplace=True)

train = df.iloc[:300]
test = df.iloc[300:]

features = [c for c in df.columns if c != 'value']
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(train[features], train['value'])
pred = model.predict(test[features])
print(f'MAE: {mean_absolute_error(test["value"], pred):.4f}')
```
