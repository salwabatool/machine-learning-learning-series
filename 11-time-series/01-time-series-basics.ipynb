## What is a Time Series?
Sequence of data points indexed in time order. Key properties: trend, seasonality, cycles, noise.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
dates = pd.date_range('2020-01-01', periods=200, freq='D')
trend = np.linspace(10, 50, 200)
seasonality = 10 * np.sin(2 * np.pi * np.arange(200) / 30)
noise = np.random.randn(200) * 2
ts = pd.Series(trend + seasonality + noise, index=dates)

fig, axes = plt.subplots(2,2, figsize=(12,8))
axes[0,0].plot(ts); axes[0,0].set_title('Full Time Series')
axes[0,1].plot(pd.Series(trend, index=dates)); axes[0,1].set_title('Trend')
axes[1,0].plot(pd.Series(seasonality, index=dates)); axes[1,0].set_title('Seasonality')
axes[1,1].plot(pd.Series(noise, index=dates)); axes[1,1].set_title('Noise')
plt.tight_layout(); plt.show()
```

## Stationarity
Many models require stationary series (constant mean/variance over time). Use Augmented Dickey-Fuller test.

```python
from statsmodels.tsa.stattools import adfuller
result = adfuller(ts)
print(f'ADF Statistic: {result[0]:.4f}')
print(f'p-value: {result[1]:.4f}')
print('Stationary!' if result[1] < 0.05 else 'Not stationary — differencing needed')
```
