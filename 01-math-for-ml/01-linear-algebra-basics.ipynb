## Vectors, Matrices, and Tensors
Linear algebra is the language of ML — data is vectors, weights are matrices.

```python
import numpy as np

# Vectors
v = np.array([1, 2, 3])
print('Vector:', v)
print('Norm (magnitude):', np.linalg.norm(v))
```

```python
# Matrix operations
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])
print('A+B:', A+B)
print('A@B (matmul):', A@B)
print('Transpose A:', A.T)
```

```python
# Eigenvalues & Eigenvectors (used in PCA)
vals, vecs = np.linalg.eig(A)
print('Eigenvalues:', vals)
print('Eigenvectors:', vecs)
```

## Why This Matters
- Matrix multiplication = forward pass in neural networks
- Eigendecomposition = PCA dimensionality reduction
- Dot product = similarity between vectors
