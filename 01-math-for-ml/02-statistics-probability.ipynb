## Descriptive Statistics

```python
import numpy as np
from scipy import stats

data = np.random.normal(50, 10, 1000)
print(f'Mean: {np.mean(data):.2f}')
print(f'Median: {np.median(data):.2f}')
print(f'Std Dev: {np.std(data):.2f}')
print(f'Variance: {np.var(data):.2f}')
```

## Probability Distributions

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(12,4))
normal_data = np.random.normal(0, 1, 1000)
uniform_data = np.random.uniform(0, 1, 1000)
axes[0].hist(normal_data, bins=40, color='steelblue', edgecolor='white')
axes[0].set_title('Normal Distribution')
axes[1].hist(uniform_data, bins=40, color='coral', edgecolor='white')
axes[1].set_title('Uniform Distribution')
plt.tight_layout()
plt.show()
```

## Bayes Theorem
P(A|B) = P(B|A) * P(A) / P(B)

This is the foundation of Naive Bayes and Bayesian ML.
