## What is Optimization in ML?
ML training = finding parameters that minimize a loss function.

```python
import numpy as np
import matplotlib.pyplot as plt

# Loss landscape
x = np.linspace(-5, 5, 200)
y = x**2 + 2*np.sin(5*x)
plt.figure(figsize=(8,4))
plt.plot(x, y)
plt.title('Example Loss Landscape')
plt.xlabel('Parameter')
plt.ylabel('Loss')
plt.show()
```

## Gradient Descent Variants
- **Batch GD**: Uses all data — stable but slow
- **Stochastic GD (SGD)**: One sample at a time — noisy but fast
- **Mini-batch GD**: Best of both — most commonly used

```python
# Comparing learning rates
def gradient_descent(lr, steps=100):
    x = 10.0
    path = [x]
    for _ in range(steps):
        x -= lr * 2 * x
        path.append(x)
    return path

plt.figure(figsize=(10,4))
for lr in [0.01, 0.1, 0.5]:
    plt.plot(gradient_descent(lr, 50), label=f'lr={lr}')
plt.title('Effect of Learning Rate')
plt.legend()
plt.show()
```
