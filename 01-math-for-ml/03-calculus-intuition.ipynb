## Derivatives — The Core of Learning
Derivatives tell us how to change weights to reduce error.

```python
import numpy as np
import matplotlib.pyplot as plt

# Simple function: f(x) = x^2
x = np.linspace(-3, 3, 100)
y = x**2
dy = 2*x  # derivative

plt.figure(figsize=(8,4))
plt.plot(x, y, label='f(x) = x²')
plt.plot(x, dy, label="f'(x) = 2x", linestyle='--')
plt.axhline(0, color='black', linewidth=0.5)
plt.legend()
plt.title('Function and Its Derivative')
plt.show()
```

## Gradient Descent Intuition

```python
# Gradient descent to minimize f(x) = x^2
x = 10.0  # starting point
lr = 0.1  # learning rate
history = []
for _ in range(50):
    gradient = 2 * x
    x = x - lr * gradient
    history.append(x)
plt.plot(history)
plt.title('Gradient Descent Convergence')
plt.xlabel('Iteration')
plt.ylabel('x value')
plt.show()
print(f'Minimum found at x ≈ {x:.6f}')
```

## Chain Rule — Backpropagation
Chain rule lets us compute gradients through layers:
d(L)/d(w) = d(L)/d(output) * d(output)/d(w)
