## What is Naive Bayes?
Probabilistic classifier based on Bayes Theorem, assuming feature independence.

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Text classification example
categories = ['sci.space', 'rec.sport.hockey', 'talk.politics.misc']
train = fetch_20newsgroups(subset='train', categories=categories)
test = fetch_20newsgroups(subset='test', categories=categories)
```

```python
vectorizer = TfidfVectorizer(max_features=5000)
X_train = vectorizer.fit_transform(train.data)
X_test = vectorizer.transform(test.data)

model = MultinomialNB()
model.fit(X_train, train.target)
y_pred = model.predict(X_test)
print(f'Accuracy: {accuracy_score(test.target, y_pred):.4f}')
```

## Summary
- Very fast and good for text classification
- 'Naive' because it assumes features are independent
- Works surprisingly well despite the strong assumption
